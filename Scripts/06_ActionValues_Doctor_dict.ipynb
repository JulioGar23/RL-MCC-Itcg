```python
# Bandits Problem - Action-Value Estimation
# Normal Script-type for the medical bottles experiment
# First version by JAGR.
# Sample average method to update action-values using dictionaries.

import random

q_values = {"P": 0.0, "Y": 0.0, "B": 0.0}
counts = {"P": 0, "Y": 0, "B": 0}
acum_rewards = {"P": 0, "Y": 0, "B": 0}

# Real World Simulation (Hidden Probabilities, Hidden Distributions)
# Hidden to the agent

real_probabilities={"P": 0.50, "Y": 0.95, "B": 0.30}

for t in range(1, 600): # iterations
    selected_action = random.choice(["P", "Y", "B"])   # Random action selection

    if random.random() < real_probabilities[selected_action]:
        reward = 1
    else:
        reward = 0
    
    counts[selected_action] += 1
    acum_rewards[selected_action] += reward

    q_values[selected_action] = acum_rewards[selected_action] / counts[selected_action]
    #print(f"Iteration {t}: Selected Bottle: {selected_action}, Reward: {reward}, Q-values: {q_values}, Acumulated Rewards: {acum_rewards}, Counts: {counts}")

# Final Q-values after all iterations
print(f"Final Q-values: {q_values}")


```
